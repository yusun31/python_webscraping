### 기상청 날씨 데이터파싱
- BeautifulSoup 객체의 find(), find_all()함수를 사용항여 파싱하기
- 파싱한 데이터를 dict, list에 저장하기
- json file로 저장하기

```python
import requests
from bs4 import BeautifulSoup

url = 'http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp'

res = requests.get(url)
if res.ok:
    soup = BeautifulSoup(res.text, 'html.parser')
    
    title_tag = soup.find('title')
    print(type(title_tag), title_tag, title_tag.text)
    
    title_all_tag = soup.find_all('title')
    for title in title_all_tag:
        print(type(title.string), type(title.text))
        #print(title.string, title.text)
        
    location_tag = soup.find('location', attrs ={'wl_ver':3})
    print(type(location_tag)) #, location_tag, location_tag.text)
    
    location_tag2 = soup.select_one("location[wl_ver='3']")
    print(type(location_tag2)) #, location_tag2, location_tag2.text)
    
    loc_tag_all = soup.findAll('location', attrs={'wl_ver':3})
    print(type(loc_tag_all))
    
    loc_tag_all2 = soup.select("location[wl_ver='3']")
    print(type(loc_tag_all2))
```

#### 하나의 지역 날씨 데이터 가져오기

```python
import requests
from bs4 import BeautifulSoup

url = 'http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp'

res = requests.get(url)
if res.ok:
    soup = BeautifulSoup(res.text, 'html.parser')
    
    location_tag = soup.find('location', attrs={'wl_ver':3})
    location_dict = {} # 초기화
    
    location_dict['province'] = location_tag.find('province').text
    location_dict['city'] = location_tag.find('city').text
    
    data_tags = location_tag.find_all('data')
    data_list = []
    for data_tag in data_tags:
        data_dict = {}
        data_dict['mode'] = data_tag.find('mode').text
        data_dict['tmef'] = data_tag.find('tmef').text
        data_dict['wf'] = data_tag.find('wf').text
        data_dict['tmn'] = data_tag.find('tmn').text
        data_dict['tmx'] = data_tag.find('tmx').text
        data_list.append(data_dict)
    location_dict['datas'] = data_list
print(location_dict)
```

#### 지역별 날씨 데이터 가져오기

```python
import requests
from bs4 import BeautifulSoup

url = 'http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp'

res = requests.get(url)
if res.ok:
    soup = BeautifulSoup(res.text, 'html.parser')
    location_tags = soup.find_all('location', attrs={'wl_ver':3})
    
    location_list = []
    for loc_tag in location_tags:
        location_dict={}
        location_dict['province'] = loc_tag.find('province').text
        location_dict['city'] = loc_tag.find('city').text
        
        data_ts = loc_tag.find_all('data')
        data_list = []
       
        for data in data_ts:
            data_dict = {}
            data_dict['mode'] = data.find('mode').text
            data_dict['tmef'] = data.find('tmef').text
            data_dict['wf'] = data.find('wf').text
            data_dict['tmn'] = data.find('tmn').text
            data_dict['tmx'] = data.find('tmx').text
            data_list.append(data_dict)
            
        location_dict['datas'] = data_list
        location_list.append(location_dict)
print(len(location_list))        
print(location_list)
```

### location_list를 json 파일로 저장하기

```python
import json

with open('data\\weather.json','w') as file:
    json.dump(location_list, file)
```

```python
with open('data\\weather.json', 'r', encoding='utf-8') as file:
    content = file.read()
    json_content = json.loads(content)
print(json_content)
```
